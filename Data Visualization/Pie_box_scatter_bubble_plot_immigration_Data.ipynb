```python
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df_can = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv')

print('Data read into a pandas dataframe!')
```

```python
df_can.head()
```

```python
print(df_can.shape)
```

```python
df_can.set_index('Country', inplace=True)
```

```python
df_can.head()
```

```python
print('data dimensions:', df_can.shape)
```

```python
years = list(map(str, range(1980, 2014)))
```

<h1> Making pie charts</h1>

```python
#Grouping data and applying sum function

df_continents = df_can.groupby('Continent', axis=0).sum()
print(type(df_can.groupby('Continent', axis=0)))
df_continents.head()
```

```python
#Creating a pie chart 

df_continents['Total'].plot(kind='pie',
                            figsize=(5, 6),
                            autopct='%1.1f%%',
                            startangle=90,     
                            shadow=True,         
                            )

plt.title('Immigration to Canada by Continent [1980 - 2013]')
plt.axis('equal')
plt.legend(labels=df_continents.index, loc='upper left') 

plt.show()
```

```python
#Improving the visual of pie chart 

colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
explode_list = [0.1, 0, 0, 0, 0.1, 0.1] 

df_continents['Total'].plot(kind='pie',
                            figsize=(10, 6),
                            autopct='%1.1f%%', 
                            startangle=90,    
                            shadow=True,       
                            labels=None,        
                            pctdistance=1.12,  
                            )

plt.title('Immigration to Canada by Continent [1980 - 2013]', y=1.12, fontsize = 15) 
plt.axis('equal') 
plt.legend(labels=df_continents.index, loc='upper left', fontsize=7) 

plt.show()
```

```python
#New immigrants grouped by continents in the year 2013

explode_list = [0.0,0,0,0.1,0.1,0.2]
df_continents['2013'].plot(kind = 'pie',
                        figsize=(15,6),
                          autopct = '%1.1f%%',
                          startangle = 90,
                          shadow = True,
                          labels = None ,
                          pctdistance = 1.12,
                          explode = explode_list
                          )

plt.title('Immigration to canada by continent 2013',y=1.12)
plt.axis('equal')
plt.legend(labels= df_continents.index, loc = 'upper left')

plt.show()
```

```python
#Creating box plot to identify immigration from top 15 countries

df_top15 = df_can.sort_values(['Total'], ascending = False , axis =0).head(15)
df_top15
```

```python
#Creating a new dataframe which has aggregate for each decade
year80 = list(map(str,range(1980,1990)))
year90 = list(map(str,range(1990,2000)))
year00 = list(map(str,range(2000,2010)))

df80 = df_top15.loc[:, year80].sum(axis=1)
df90 = df_top15.loc[:, year90].sum(axis=1)
df00 = df_top15.loc[:, year00].sum(axis=1)

new_df= pd.DataFrame({'1980s':df80 , '1990s':df90,'2000s':df00})
new_df.head()
```

<h1>Scatter plots</h1>

```python
#Plotting trend of overall immigration to canada(from all countries) since 1980

df_tot = pd.DataFrame(df_can[years].sum(axis=0))

df_tot.index = map(int, df_tot.index)

df_tot.reset_index(inplace = True)

df_tot.columns = ['year', 'total']

df_tot.head()
```

```python
df_tot.plot(kind='scatter', x='year', y='total', figsize=(10, 6), color='darkblue')

plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

plt.show()
```

```python
#Adding a regression line
x = df_tot['year']   
y = df_tot['total']    
fit = np.polyfit(x, y, deg=1)

fit
```

```python
df_tot.plot(kind='scatter', x='year', y='total', figsize=(10, 6), color='darkblue')

plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

plt.plot(x, fit[0] * x + fit[1], color='red')
plt.annotate('y={0:.0f} x + {1:.0f}'.format(fit[0], fit[1]), xy=(2000, 150000))

plt.show()

'No. Immigrants = {0:.0f} * Year + {1:.0f}'.format(fit[0], fit[1]) 
```

```python
#Creating scatter plot of total immigration from denmark,norway,sweden

df_countries = df_can.loc[['Denmark','Norway','Sweden'],years].transpose()

df_total = pd.DataFrame(df_countries.sum(axis=1))

df_total.reset_index(inplace = True)

df_total.columns = ['year','total']

df_total['year'] = df_total['year'].astype(int)

df_total.head()
```

```python
#Sctter plot of total vs year

df_total.plot(kind = 'scatter',
              x = 'year',
              y='total',
             figsize = (10,6),color = 'purple')

plt.title('Immigration from Denmark, Norway, and Sweden to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.show()
```

<h1>Bubble plot</h1>

```python
#Comparing argentina vs brazil immigration

df_can_t = df_can[years].transpose()

df_can_t.index = map(int, df_can_t.index)

df_can_t.index.name = 'Year'

df_can_t.reset_index(inplace=True)

df_can_t.head()
```

```python
#Normalizing Brazil and argentina data

norm_brazil = (df_can_t['Brazil'] - df_can_t['Brazil'].min()) / (df_can_t['Brazil'].max() - df_can_t['Brazil'].min())

norm_argentina = (df_can_t['Argentina'] - df_can_t['Argentina'].min()) / (df_can_t['Argentina'].max() - df_can_t['Argentina'].min())
```

```python
#Plotting graph

ax0 = df_can_t.plot(kind='scatter',
                    x='Year',
                    y='Brazil',
                    figsize=(14, 8),
                    alpha=0.5, 
                    color='green',
                    s=norm_brazil * 2000 + 10, 
                    xlim=(1975, 2015)
                    )


ax1 = df_can_t.plot(kind='scatter',
                    x='Year',
                    y='Argentina',
                    alpha=0.5,
                    color="blue",
                    s=norm_argentina * 2000 + 10,
                    ax=ax0
                    )

ax0.set_ylabel('Number of Immigrants')
ax0.set_title('Immigration from Brazil and Argentina from 1980 to 2013')
ax0.legend(['Brazil', 'Argentina'], loc='upper left', fontsize='x-large')
```

```python
#Plotting bubble graph for china vs india

#Normalizing data
norm_china =  (df_can_t['China'] - df_can_t['China'].min())/ (df_can_t['China'].max() - df_can_t['China'].min())

norm_india= (df_can_t['India'] - df_can_t['India'].min())/ (df_can_t['India'].max() - df_can_t['India'].min())
```

```python
ax0 = df_can_t.plot(kind = 'scatter',
                    x = 'Year',
                    y= 'China',
                    figsize=(14,8),
                    alpha = 0.5,
                    color = 'green',
                    s = norm_china *2000+10,
                    xlim = (1975,2015)
                   )

ax1 = df_can_t.plot(kind='scatter',
                    x='Year',
                    y='India',
                    alpha=0.5,
                    color="blue",
                    s=norm_india * 2000 + 10,
                    ax = ax0
                    )

ax0.set_ylabel('Number of Immigrants')
ax0.set_title('Immigration from China and India from 1980 - 2013')
ax0.legend(['China', 'India'], loc='upper left', fontsize='x-large')
```
