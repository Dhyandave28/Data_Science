<h1>Bubble Plot</h1>

```python
#Creating bubble plot parameters

crime_details={
    'city': ['Chicago', 'Chicago','Austin','Austin','Seattle','Seattle'],
    'Numberofcrimes': [1000,1200,400,700,350,1500],
    'Year': ['2007','2008','2007','2008','2007','2008'],
}

df = pd.DataFrame(crime_details)

df
```

```python
#Grouping data for bubble plot
bub_data = df.groupby('city')['Numberofcrimes'].sum().reset_index()
```

```python
bub_data
```

```python
#Plot
fig = px.scatter(bub_data,x ="city", y ="Numberofcrimes", size = "Numberofcrimes",
                hover_name = "city", title ="Crime statistics" , size_max=60
                )

fig.show()
```

<h1>Sunburst chart</h1>

```python
#Sunburst parameter and giving value to prameter

data = dict(
    characters=["Eve", "Cain", "Seth", "Enos", "Noam", "Abel", "Awan", "Enoch", "Azura"],
    parent=["",'Eve',"Eve","Seth","Seth","Eve","Eve","Awan","Eve"],
    value =[10,14,12,10,2,6,6,4,4]
)

fig = px.sunburst(
    data,
    names ='characters',
    parents = 'parent',
    values = 'value',
    title = "Family chart"
)

fig.show()
```

<h1>Plotly skills to Airline dataset</h1>

```python
#Importing data for plots

import io
import requests

url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/airline_data.csv'
resp = requests.get(url)
text = io.BytesIO(resp.content)

airline_data = pd.read_csv( text,
                           encoding = 'ISO-8859-1',
                           dtype={'DIV1Airpot':str , 'Div1TailNum': str,
                                 'DIV2Airport':str, 'Div2TailNum':str
                                 })

print('Data downloaded')
    
```

```python
airline_data.head()
```

```python
airline_data.shape
```

```python
data = airline_data.sample(n=500,random_state=42)
```

```python
data.shape
```

<h2>Scatter plot</h2>

```python
#Creating empty chart
fig = go.Figure()
fig
```

```python
#Fianl plot
fig.add_trace(go.Scatter(x=data['Distance'], y= data['DepTime'], mode = 'markers',marker = dict(color = 'red')))
fig.update_layout(title = 'Distance vs DepTime', xaxis_title ='Distance',yaxis_title ='DepTime')
fig.show()
```

<h2>Line plot</h2>

```python
line_data = data.groupby('Month')['ArrDelay'].mean().reset_index()
```

```python
line_data
```

```python
#Empty chart
fig = go.Figure()
fig
```

```python
#Final plot
fig.add_trace(go.Scatter(x=line_data['Month'], y = line_data['ArrDelay'],mode = 'lines',marker=dict(color = 'green')))
fig.update_layout(title='Month vs Avg fligh delay time', xaxis_title = "Month", yaxis_title='Arr Delay')
fig.show()
```
